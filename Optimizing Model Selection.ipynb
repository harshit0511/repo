```python
def cleanBosonData(infile_name):
    
    ##### Reads Data into Pandas DataFrame #####
    df = pd.read_csv(infile_name)

    ##### Convert feature label to numeric #####
            
    # Creating new DataFrame for the new column to store numerical values of 'Label'
    N_Label=pd.DataFrame()
    for i in df['Label']:
        if i == 's':
            N_Label=N_Label.append([1], ignore_index = True)
        else:
            N_Label=N_Label.append([0], ignore_index = True)
        
    # Concatenating new DataFrame with the existing one (df)
    frames=[df, N_Label]
    df=pd.concat(frames, axis=1)
    df['N_Label']=df[0]
    df=df.drop(['Label', 0], axis=1)
    
    ##### Creating dummy variable for missing values
    
    DER_mass_MMC_mv=pd.DataFrame()
    for i in df['DER_mass_MMC']:
        if i == -999:
            DER_mass_MMC_mv=DER_mass_MMC_mv.append([1], ignore_index = True)
        else:
            DER_mass_MMC_mv=DER_mass_MMC_mv.append([0], ignore_index = True)

    DER_deltaeta_jet_jet_mv=pd.DataFrame()
    for i in df['DER_deltaeta_jet_jet']:
        if i == -999:
            DER_deltaeta_jet_jet_mv=DER_deltaeta_jet_jet_mv.append([1], ignore_index = True)
        else:
            DER_deltaeta_jet_jet_mv=DER_deltaeta_jet_jet_mv.append([0], ignore_index = True)

    DER_mass_jet_jet_mv=pd.DataFrame()
    for i in df['DER_mass_jet_jet']:
        if i == -999:
            DER_mass_jet_jet_mv=DER_mass_jet_jet_mv.append([1], ignore_index = True)
        else:
            DER_mass_jet_jet_mv=DER_mass_jet_jet_mv.append([0], ignore_index = True)

    DER_prodeta_jet_jet_mv=pd.DataFrame()
    for i in df['DER_prodeta_jet_jet']:
        if i == -999:
            DER_prodeta_jet_jet_mv=DER_prodeta_jet_jet_mv.append([1], ignore_index = True)
        else:
            DER_prodeta_jet_jet_mv=DER_prodeta_jet_jet_mv.append([0], ignore_index = True)

    # Joining the DataFrames and renaming the columns
    
    df=pd.concat([df,DER_mass_MMC_mv], axis=1)
    df['DER_mass_MMC_mv']=df[0]
    df=df.drop([0], axis=1)
    
    df=pd.concat([df,DER_deltaeta_jet_jet_mv], axis=1)
    df['DER_deltaeta_jet_jet_mv']=df[0]
    df=df.drop([0], axis=1)

    df=pd.concat([df,DER_mass_jet_jet_mv], axis=1)
    df['DER_mass_jet_jet_mv']=df[0]
    df=df.drop([0], axis=1)

    df=pd.concat([df,DER_prodeta_jet_jet_mv], axis=1)
    df['DER_prodeta_jet_jet_mv']=df[0]
    df=df.drop([0], axis=1)
    
    ##### Setting the Index to 'EventId' column
    df.set_index('EventId', inplace=True)
    
    ##### Replacing -999 with NaN
    for col in df:
        df[col] = df[col].replace(-999, np.nan)
    
    ##### Replacing NaN values with the mean of that feature
    df=df.fillna(df.mean())
    
    ##### Re-scaling all the values using StandardScaler() from sklearn.preprocessing
    column=['DER_mass_MMC', 'DER_mass_transverse_met_lep', 'DER_mass_vis', 'DER_pt_h', 'DER_deltaeta_jet_jet', 'DER_mass_jet_jet', 'DER_prodeta_jet_jet', 'DER_deltar_tau_lep', 'DER_pt_tot', 'DER_mass_MMC_mv', 'DER_deltaeta_jet_jet_mv', 'DER_mass_jet_jet_mv', 'DER_prodeta_jet_jet_mv']
    for i in column:
        df[i]=scale(df[i])
    column_orig=['DER_mass_MMC', 'DER_mass_transverse_met_lep', 'DER_mass_vis', 'DER_pt_h', 'DER_deltaeta_jet_jet', 'DER_mass_jet_jet', 'DER_prodeta_jet_jet', 'DER_deltar_tau_lep', 'DER_pt_tot', 'N_Label', 'DER_mass_MMC_mv', 'DER_deltaeta_jet_jet_mv', 'DER_mass_jet_jet_mv', 'DER_prodeta_jet_jet_mv']
    data_clean=pd.DataFrame(df, columns=column_orig)
    
    return data_clean
```
